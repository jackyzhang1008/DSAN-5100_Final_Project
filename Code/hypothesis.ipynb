Compare the average duration between 'Member' and 'Casual' users in your dataset on precovid time

$H_o$: There is no significant difference in the average duration between 'Member' and 'Casual' users.

$H_a$:  There is a significant difference in the average duration between 'Member' and 'Casual' users.

Test Type: T-Test

Decision: If p-value is smaller than $\alpha=0.05$ significance level, we reject $H_o$. If p=value is greater than $\alpha=0.05$ significance level, then we do not reject $H_o$.


```python
#precovid
member_durations = data[data["Member type"] == "Member"]["Duration"]
casual_durations = data[data["Member type"] == "Casual"]["Duration"]

# Performing a t-test to compare the means
t_stat, p_value = stats.ttest_ind(member_durations, casual_durations, equal_var=False)

t_stat, p_value
```

Conclusion: Since p-value from t-test is 0.0, we have enough evidence to reject the null hypothesis $\alpha=0.05$ significance level. Thus, there is significant difference in the average duration between 'Member' and 'Casual' users.

Compare 'Start station number' and 'Member type' precovid

$H_o$: There is no significant association between 'Start station number' and 'Member type'.

$H_a$: There is a significant association between 'Start station number' and 'Member type'.

Test Type: Chi-Squared Test

Decision: If p-value is smaller than $\alpha=0.05$ significance level, we reject $H_o$. If p=value is greater than $\alpha=0.05$ significance level, then we do not reject $H_o$.


```python
from scipy.stats import chi2_contingency
```

```python
contingency_table = pd.crosstab(data['Start station number'], data['Member type'])

chi2, p, dof, expected = chi2_contingency(contingency_table)

chi2, p, dof, expected
```

Conclusion: Since p-value from chi-squared test is 0.0, we have enough evidence to reject the null hypothesis $\alpha=0.05$ significance level. Thus, there is significant association between 'Start station number' and 'Member type'.
